```python
import pandas as pd
import numpy as np

df = pd.read_excel(
    'indo_12_1.xls', 
    skiprows=3, 
    skipfooter=2, 
    na_values = ['-'] #untuk melihat data yang kosong otomatis
)
df.rename(columns={'Unnamed: 0':'Provinsi'}, inplace=True)
df
```

```python
year = list(df.columns[1:])
year
```

```python
min = df[df[1971] == df[1971].min()]
# min
listMin = list(min.values[0])[1:]
listMin
```

```python
max = df[df[2010] == df[2010][:-1].max()]
# max
listMax = list(max.values[0])[1:]
listMax
```

```python
ind = list(df.iloc[-1])[1:]
ind
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(year, ind, 'r--', year, listMin, 'b--', year, listMax, 'g--')
plt.show()
```
